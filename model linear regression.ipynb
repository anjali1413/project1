# feature engineering 

```python
# plt.scatter(data["Sales_in_thousands"],data["Model"])
```

```python
from datetime import date
data['Latest_Launch'] = pd.to_datetime(data['Latest_Launch'])
data["age_car"] = date.today().year - data['Latest_Launch'].dt.year
```

```python
data["Area_car"] = data["Length"]+data["Width"]
```

```python
data.isnull().sum()
```

```python
data.drop(["four_year_resale_value"],axis=1,inplace=True)
```

```python
cat_data = data[data.dtypes[data.dtypes=="object"].index]
num_data=data[data.dtypes[data.dtypes!="object"].index]
```

```python
cat_data.isnull().sum()
```

```python
num_data.isnull().sum()
```

```python
num_data.fillna(num_data.median(),inplace = True)
```

```python
num_data.isnull().sum()
```

```python
# outlier treatment 
```

```python
data.describe(percentiles=[0.01,0.05,0.07,0.10,0.25,0.45,0.50,0.60,0.70,0.75,0.95,0.06,0.97,0.98,0.99]).T
# from the data we will see that outlier is present 
```

```python
# treating outlier 
def outlier(x):
    x = x.clip(upper=x.quantile(0.9))
    return x
```

```python
num_data = num_data.apply(outlier)
```

```python
num_data.describe(percentiles=[0.01,0.05,0.20,0.50,0.60,0.75,.80,0.95,0.96,0.97,0.98,0.99]).T
```

```python
final_data = pd.concat([cat_data,num_data],axis = 1)
```

```python
# now converting our data into non numerical to numerical 
```

```python
final_data0 = pd.get_dummies(final_data,columns=["Manufacturer","Vehicle_type"],drop_first=True)
```

```python
final_data0.columns
```

```python
# now checking multicolineariy 
```

```python
cor = final_data0[['Model', 'Sales_in_thousands', 'Price_in_thousands', 'Engine_size',
       'Horsepower', 'Wheelbase', 'Width', 'Length', 'Curb_weight',
       'Fuel_capacity', 'Fuel_efficiency', 'Latest_Launch',
       'Power_perf_factor']].corr()
```

```python
sns.heatmap(cor,annot=True,cmap="coolwarm")
```

```python
# Autocorrelation (residual/error term - predicted value - actual value )
# darvin watson test 
# ho : there is no correaltion among the residual 
# ha : there is correlation 
# CI = 0.95 , alpha = 0.05 

# decision p> alpha :- accept Ha
#         p< alpha : reject ha
```

```python

plt.figure(figsize=(5,3))
sns.histplot(final_data0["Sales_in_thousands"])
plt.show()
```

```python
# checking data is normally distributed or not 
```

```python
from scipy import stats 
stats.shapiro(final_data0["Sales_in_thousands"])
```

```python
pvalue=3.782964566600988e-11

if pvalue>0.05:
    print("ho is accepted->data is normally distributed")
else:
    print("h1 is accepted -> data is not normally distributed")
```

```python
# by seeing the result we see that our data is not normally distributed so changing it into normal 
```

```python
final_data0["Sales"] = np.log1p(final_data0["Sales_in_thousands"])

print(stats.shapiro(final_data0["Sales"]))
plt.figure(figsize=(5,3))
sns.histplot(final_data0["Sales"])
plt.show()
```

```python
# Model Development 
```

```python
y = final_data0["Sales"]
x = final_data0[[ 'Price_in_thousands', 'Engine_size',
       'Horsepower', 'Wheelbase', 'Width', 'Length', 'Curb_weight',
       'Fuel_capacity', 'Fuel_efficiency', 
       'Power_perf_factor', 'age_car', 'Area_car', 'Manufacturer_Audi',
       'Manufacturer_BMW', 'Manufacturer_Buick', 'Manufacturer_Cadillac',
       'Manufacturer_Chevrolet', 'Manufacturer_Chrysler', 'Manufacturer_Dodge',
       'Manufacturer_Ford', 'Manufacturer_Honda', 'Manufacturer_Hyundai',
       'Manufacturer_Infiniti', 'Manufacturer_Jaguar', 'Manufacturer_Jeep',
       'Manufacturer_Lexus', 'Manufacturer_Lincoln', 'Manufacturer_Mercedes-B',
       'Manufacturer_Mercury', 'Manufacturer_Mitsubishi',
       'Manufacturer_Nissan', 'Manufacturer_Oldsmobile',
       'Manufacturer_Plymouth', 'Manufacturer_Pontiac', 'Manufacturer_Porsche',
       'Manufacturer_Saab', 'Manufacturer_Saturn', 'Manufacturer_Subaru',
       'Manufacturer_Toyota', 'Manufacturer_Volkswagen', 'Manufacturer_Volvo',
       'Vehicle_type_Passenger']]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=99,test_size=0.3)
```

```python
import statsmodels.api as sm 
x_train1= x_train[[ 'Price_in_thousands', 'Engine_size',
       'Horsepower', 'Wheelbase', 'Width', 'Length', 'Curb_weight',
       'Fuel_capacity', 'Fuel_efficiency', 
       'Power_perf_factor', 'age_car', 'Area_car', 'Manufacturer_Audi',
       'Manufacturer_BMW', 'Manufacturer_Buick', 'Manufacturer_Cadillac',
       'Manufacturer_Chevrolet', 'Manufacturer_Chrysler', 'Manufacturer_Dodge',
       'Manufacturer_Ford', 'Manufacturer_Honda', 'Manufacturer_Hyundai',
       'Manufacturer_Infiniti', 'Manufacturer_Jaguar', 'Manufacturer_Jeep',
       'Manufacturer_Lexus', 'Manufacturer_Lincoln', 'Manufacturer_Mercedes-B',
       'Manufacturer_Mercury', 'Manufacturer_Mitsubishi',
       'Manufacturer_Nissan', 'Manufacturer_Oldsmobile',
       'Manufacturer_Plymouth', 'Manufacturer_Pontiac', 'Manufacturer_Porsche',
       'Manufacturer_Saab', 'Manufacturer_Saturn', 'Manufacturer_Subaru',
       'Manufacturer_Toyota', 'Manufacturer_Volkswagen', 'Manufacturer_Volvo',
       'Vehicle_type_Passenger']]

x_train1 = sm.add_constant(x_train1)
model= sm.OLS(y_train,x_train1)
results=model.fit()
print(results.summary())
```
