# Exploratory data analysis

```python
data.columns
```

### Gender

```python
plt.figure(figsize=(5,4))
ax = sns.countplot(x='Gender',data=data)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
data.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by=['Amount'], ascending=False)
```

```python
sales_gen = data.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by=['Amount'], ascending=False)
sns.barplot(x='Gender',y='Amount',data=sales_gen)
```

```python
# from the above visualization we can see that our most of the buyers are female and 
# even the purchasing power of female our greater 
```

### AGE

```python
ax = sns.countplot(data=data,x='Age Group',hue='Gender')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
data.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by=['Amount'], ascending=False)
```

```python
sales_age = data.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by=['Amount'], ascending=False)

sns.barplot(x='Age Group',y='Amount',data=sales_age)
```

```python
# from the above visualization we can see that our most of the buyers are of age 26-35 and mostly females 
```

### State

```python
# total number of orders for top 10 states 
```

```python
sales_state = data.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sales_state
```

```python
plt.figure(figsize=(15,5))
sns.barplot(x='State',y='Orders',data=sales_state)
```

```python
sales_state = data.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
plt.figure(figsize=(15,5))
sns.barplot(x='State',y='Amount',data=sales_state)
```

```python
# from the above visualoization we can see that our top states are up , mahrashtra and so on...
# and same total purchased the only diff is of kerala and haryana... kerala has high in orders
# but haryana has high purchased amount 
```

### Marital status 

```python
ax=sns.countplot(data=data,x='Marital_Status')
plt.figure(figsize=(2,3))
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = data.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
# plt.figure(figsize=(15,5))
sns.barplot(x='Marital_Status',y='Amount',data=sales_state,hue='Gender')
```

```python
# from the above visualization we can see that most of the buyer are married(women) and they have high purchasing power 
```

### Occupation

```python
plt.figure(figsize=(15,5))
ax=sns.countplot(data=data,x='Occupation')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = data.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
plt.figure(figsize=(15,5))
sns.barplot(x='Occupation',y='Amount',data=sales_state)
```

```python
# from the above graph we can see that most of the buyer are working in IT Sector , Healthcare and then aviation ....
# and so on
```

### Product Category 

```python
plt.figure(figsize=(25,10))
ax=sns.countplot(data=data,x='Product_Category')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = data.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',
ascending=False).head(10)


plt.figure(figsize=(20,5))
sns.barplot(x='Product_Category',y='Amount',data=sales_state)
```

```python
# from the above graph we can see that most of the sold products are from food, footwear, and electronic category 
```

### Product ID

```python
sales_state = data.groupby(['Product_ID'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
plt.figure(figsize=(15,5))
sns.barplot(x='Product_ID',y='Amount',data=sales_state)
```
